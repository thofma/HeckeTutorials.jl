```julia
Zx, x = FlintZZ["x"]
C = MSet{fmpz}()
U = MSet{fmpz}()
@time for i=2:1000
    I = i+10^10
    if all(isone, values(factor(I).fac))
        k,a = number_field(x^2-I)
        zk = maximal_order(k)
        c, mc = class_group(zk)
        u, mu = unit_group_fac_elem(zk)
        push!(C, order(c))
        push!(U, norm(mu(u[2])))
    end
end
println(C)
println(U)
```
